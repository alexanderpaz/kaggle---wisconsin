# Carga de librerías y setup inicial

```python
import pandas as pd
import plotnine as p9
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Configuramos pandas para que nos muestre todas las columnas

pd.set_option('display.max_columns', None)
```

# Carga de datos

```python
# Cargamos los datos desde archivo csv a Pandas

ds = pd.read_csv('../raw/data-wisconsin.csv')
```

```python
# Visualizamos la estructura de los datos

ds.info()
```

# Preparación de la data: transformaciones

```python
# Mostramos los primeros registros para tener una idea de la data

ds.head()
```

```python
# Nos damos cuenta que existen dos columnas que podemos prescindir para nuestro análisis y desarrollo del modelo: id y unnamed: 32

ds = ds.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
# Vamos a agregar una columna numérica que nos indique si el paciente tiene o no cáncer, esto para poder realizar correlaciones y visualizaciones

ds['Diagnóstico'] = ds['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Vamos a reemplazar M y B por Maligno y Benigno para que sea más fácil de entender

ds['diagnosis'] = ds['diagnosis'].map({'M': 'Maligno', 'B': 'Benigno'})
```

# EDA: Exploratory Data Analysis

```python
# Seguimos conociendo la data y para ello obtenemos las estadísticas descriptivas más básicas

ds.describe()
```

```python
# Vamos a ver la distribución de los datos para cada una de las variables

ds.drop('Diagnóstico', axis=1).hist(figsize=(30, 30));
```

```python
# Otra visualización muy utilizada es el boxplot. Aquí utilizamos a pandas directamente para generar los boxplots

ds.drop('Diagnóstico', axis=1).plot(kind='box', subplots=True, layout=(6, 6), sharex=False, sharey=False, figsize=(30, 30));
```

```python
# Graficamos la distribución del Diagnóstico para ver si los datos están balanceados

sns.countplot(data = ds, x = 'diagnosis');
```

```python
# Graficamos una matriz de correlación para los valores medios

ds_medios = ds[ds.columns[ds.columns.str.contains('mean')]]

# concatenamos ds_medios con Diagnóstico

ds_medios = pd.concat([ds_medios, ds['Diagnóstico']], axis=1)
```

```python
# Encontramos la correlación entre las variables y Diagnóstico

corr = ds_medios.corr()
corr['Diagnóstico'].sort_values(ascending=False)
```

```python
# Graficamos la matriz de correlación
fig = plt.figure(figsize=(15, 15))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm');
```

```python
# Del heatmap anterior podemos ver que hay variables que están muy correlacionadas entre sí, por lo que podemos eliminar algunas de ellas para evitar problemas de multicolinealidad
# Candidatas son: radius_mean, area_mean
# Por otro lado, vemos que fractal_dimension_mean no está muy correlacionada con Diagnóstico, por lo que podemos eliminarla también

ds_medios = ds_medios.drop(['radius_mean', 'area_mean', 'fractal_dimension_mean'], axis=1)
```

```python
# Guardamos el dataset con los valores medios en un archivo csv

ds_medios.to_csv('../dataset/dataset_wisconsin.csv', index=False)
```
